###     **2. Data Preprocessing**

#### **i) Data Cleaning**

#####     **a. Clean Column Names**

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

#####     **b. Handle Missing Values**

```python
df.isnull().sum()
```

#####     **c. Clean Textual Data**

```python
df['employment_type'] = df['employment_type'].str.lower().str.strip()
df['experience_level'] = df['experience_level'].str.upper().str.strip()
df['industry'] = df['industry'].str.title().str.strip()
df['education_required'] = df['education_required'].str.title().str.strip()
```

#### **ii) Data Transformation**

```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}

# Encode all object (string) columns except 'required_skills'
for col in df.select_dtypes(include='object').columns:
    if col != 'required_skills':
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        label_encoders[col] = le
```

#### **iii) Feature Engineering**

```python
# List of Skills
df['skill_list'] = df['required_skills'].str.lower().str.strip().str.split(',\s*')
#get number of skills listed:
df['num_skills'] = df['skill_list'].apply(len)
#Convert Dates to Datetime format
df['posting_date'] = pd.to_datetime(df['posting_date'])
df['application_deadline'] = pd.to_datetime(df['application_deadline'])
```

###     **4. Model Building & Evaluation**

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.linear_model import BayesianRidge


from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

```

```python
def evaluate_model(name, model, X_train, X_test, y_train, y_test, results):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    print(f"\n{name}")
    print(f"MAE  : {mae:.2f}")
    print(f"RMSE : {rmse:.2f}")
    print(f"R    : {r2:.4f}")

    results[name] = [mae, rmse, r2]
```

```python
results = {}

# 1. Linear Regression
evaluate_model("Linear Regression", LinearRegression(), X_train, X_test, y_train, y_test, results)

# 2. Ridge Regression
evaluate_model("Ridge Regression", Ridge(alpha=1.0), X_train, X_test, y_train, y_test, results)

# 3. Lasso Regression
evaluate_model("Lasso Regression", Lasso(alpha=0.1), X_train, X_test, y_train, y_test, results)

# 4. Decision Tree
evaluate_model("Decision Tree", DecisionTreeRegressor(random_state=42), X_train, X_test, y_train, y_test, results)

# 5. Random Forest
evaluate_model("Random Forest", RandomForestRegressor(n_estimators=100, random_state=42), X_train, X_test, y_train, y_test, results)

# 6. Gradient Boosting
evaluate_model("Gradient Boosting", GradientBoostingRegressor(random_state=42), X_train, X_test, y_train, y_test, results)

# 7. XGBoost
evaluate_model("XGBoost", XGBRegressor(n_estimators=100, random_state=42, verbosity=0), X_train, X_test, y_train, y_test, results)

# 8. Support Vector Regression
evaluate_model("SVR", SVR(kernel='rbf'), X_train, X_test, y_train, y_test, results)

# 9. AdaBoost Regressor
evaluate_model("AdaBoost", AdaBoostRegressor(random_state=42), X_train, X_test, y_train, y_test, results)

# 10. Bayesian Ridge Regression
evaluate_model("Bayesian Ridge", BayesianRidge(), X_train, X_test, y_train, y_test, results)
```

###     **5. Model Comparison**

```python
result_df = pd.DataFrame(results, index=['MAE', 'RMSE', 'R2']).T
result_df = result_df.sort_values(by='R2', ascending=False)
print("\n Model Comparison:\n")
print(result_df)

# Visualize R2 scores
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.barplot(x=result_df.index, y='R2', data=result_df,hue='R2',palette='crest')
plt.xticks(rotation=45)
plt.title("Model Comparison by R2 Score")
plt.ylabel("R Score")
plt.tight_layout()
plt.show()
```

###     **6. Hyperparameter Tuning (Top 3 Models)**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
```

```python
# Random Forest Hyperparameters
rf_params = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt','log2']
}

rf_model = RandomForestRegressor(random_state=42)

rf_grid = GridSearchCV(rf_model, rf_params, scoring='r2', cv=3, n_jobs=-1, verbose=1)
rf_grid.fit(X_train, y_train)

print(" Random Forest Best Parameters:")
print(rf_grid.best_params_)
print(" Best R Score (Train CV):", rf_grid.best_score_)

# Evaluation on Test Set
rf_best = rf_grid.best_estimator_
rf_pred = rf_best.predict(X_test)

print("\n Random Forest Test Set Evaluation:")
print(f"MAE  : {mean_absolute_error(y_test, rf_pred):.2f}")
print(f"RMSE : {np.sqrt(mean_squared_error(y_test, rf_pred)):.2f}")
print(f"R    : {r2_score(y_test, rf_pred):.4f}")
```

```python
# Gradient Boosting Hyperparameters
gb_params = {
    'n_estimators': [100, 200],
    'learning_rate': [0.05, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'subsample': [0.8, 1.0],
    'min_samples_split': [2, 5]
}

gb_model = GradientBoostingRegressor(random_state=42)

gb_grid = GridSearchCV(gb_model, gb_params, scoring='r2', cv=3, n_jobs=-1, verbose=1)
gb_grid.fit(X_train, y_train)

print("\n Gradient Boosting Best Parameters:")
print(gb_grid.best_params_)
print(" Best R Score (Train CV):", gb_grid.best_score_)

# Evaluation on Test Set
gb_best = gb_grid.best_estimator_
gb_pred = gb_best.predict(X_test)

print("\n Gradient Boosting Test Set Evaluation:")
print(f"MAE  : {mean_absolute_error(y_test, gb_pred):.2f}")
print(f"RMSE : {np.sqrt(mean_squared_error(y_test, gb_pred)):.2f}")
print(f"R    : {r2_score(y_test, gb_pred):.4f}")
```

```python
# XGBoost Hyperparameters
xgb_params = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.05, 0.1, 0.2],
    'subsample': [0.8, 1]
}

xgb_model = XGBRegressor(random_state=42, verbosity=0)

xgb_grid = GridSearchCV(xgb_model, xgb_params, scoring='r2', cv=3, n_jobs=-1, verbose=1)
xgb_grid.fit(X_train, y_train)

print("\n XGBoost Best Parameters:")
print(xgb_grid.best_params_)
print(" Best R Score (Train CV):", xgb_grid.best_score_)

# Evaluation on Test Set
xgb_best = xgb_grid.best_estimator_
xgb_pred = xgb_best.predict(X_test)

print("\n XGBoost Test Set Evaluation:")
print(f"MAE  : {mean_absolute_error(y_test, xgb_pred):.2f}")
print(f"RMSE : {np.sqrt(mean_squared_error(y_test, xgb_pred)):.2f}")
print(f"R    : {r2_score(y_test, xgb_pred):.4f}")
```
